```python
import geonest
```

```python
m = geonest.Map(center=(37.562984, -122.514426), zoom=17)
m
```

```python
video_url = "https://static-assets.mapbox.com/mapbox-gl-js/drone.mp4"
video_bounds = ((37.56238816, -122.515963), (37.563391708, -122.5130939))
m.add_video(video_url, video_bounds)
```

```python
gif_url = "https://i.gifer.com/4j.gif"
gif_bounds = ((13, -150), (40, -120))
m.add_image(gif_url, bounds=gif_bounds)
```

```python
raster_url = "https://github.com/opengeos/datasets/releases/download/raster/LC09_039035_20240708_90m.tif"
m.add_raster(
    raster_url,
    indexes=[5, 4, 3],
    name="landsat"
)
```
