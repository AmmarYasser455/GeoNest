```python
from geonest import Map
from ipyleaflet import SplitMapControl
```

```python
m = Map(center=[30.0, 31.0], zoom=8)
m.add_basemap("Esri.WorldImagery")
m.add_basemap("OpenTopoMap")
```

```python
# Create split map control with left and right layers
control = SplitMapControl(left_layer=m.layers[-2], right_layer=m.layers[-1])
m.add_control(control)
m
```
